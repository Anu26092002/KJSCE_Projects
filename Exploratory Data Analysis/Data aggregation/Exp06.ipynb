```python
def minmaxNormalization(data,new_min,new_max):
    min_data = min(data)
    max_data = max(data)
    normData = []
    for i in range(len(data)):
        normData.append((data[i] - min_data)*(new_max - new_min)/(max_data - min_data) + new_min)
    return normData
```

```python
new_min = int(input('Enter the new minimum value : '))
new_max = int(input('Enter the new maximum value : '))
normData = minmaxNormalization(fish_data['Weight'],new_min,new_max)

table = [
    [i for i in fish_data['Weight']],
    [i for i in normData]
]

table = np.array(table)
table = np.transpose(table)

normData = pd.DataFrame(table, columns = ['Actual Fish Weight', 'Min-max Normalized Fish Weight'])
normData
```

```python
def zscoreNormalization(data):
    mean = sum(data)/len(data)   # mean
    var  = sum(pow(x-mean,2) for x in data)/len(data)  # variance
    std  = math.sqrt(var)  # standard deviation
    normData = []
    for i in range(len(data)):
        normData.append((data[i] - mean)/std)
    return normData
```

```python
normData = zscoreNormalization(fish_data['Height'])

table = [
    [i for i in fish_data['Weight']],
    [i for i in normData]
]

table = np.array(table)
table = np.transpose(table)

normData = pd.DataFrame(table, columns = ['Actual Fish Height', 'Z Score Normalized Fish Height'])
normData
```

```python
def decimalScaling(data):
    p = max([abs(int(i)) for i in data])
    j = len(str(p))
    normData = []
    for i in range(len(data)):
        normData.append((data[i])/(10**j))
    return normData
```

```python
normData = decimalScaling(fish_data['Width'])

table = [
    [i for i in fish_data['Width']],
    [i for i in normData]
]

table = np.array(table)
table = np.transpose(table)

normData = pd.DataFrame(table, columns = ['Actual Fish Width', 'Decimal Scaling Normalized Fish Width'])
normData
```

<h3>2.	Identify the suitable attributes to apply the K-means discretization techniques and display the output of discretized data</h3>

```python
dataset = fish_data.drop(['Length1','Length2','Length3','Width'],axis = 1)
dataset
```

```python
np.random.seed(200)

# number of clusters
k = 3

centroids = {
    i+1: [np.random.randint(int(min(fish_data['Height'])),int(max(fish_data['Height']))),np.random.randint(int(min(fish_data['Weight'])),int(max(fish_data['Weight'])))]
    for i in range(k)
}

fig = plt.figure(figsize=(5,5))
plt.scatter(dataset['Height'],dataset['Weight'],color='k')
colmap = {1:'r',2:'g',3:'b'}

for i in centroids.keys():
    plt.scatter(*centroids[i],color=colmap[i])
    
plt.xlim(int(min(fish_data['Height'])),int(max(fish_data['Height'])))
plt.ylim(int(min(fish_data['Weight'])),int(max(fish_data['Weight'])))
plt.show()
```

```python
# Assignment Stage

def assignment(dataset, centroids):
    
    # Calculating the Euclidean Distance
    for i in centroids.keys():
        dataset['Distance From {}'.format(i)] = (
            np.sqrt(
                  (dataset['Height'] - centroids[i][0])**2
                + (dataset['Weight'] - centroids[i][1])**2
            )
        )
        
    centroid_distance_cols = ['Distance From {}'.format(i) for i in centroids.keys()]
    dataset['Closest'] = dataset.loc[:, centroid_distance_cols].idxmin(axis = 1)
    dataset['Closest'] = dataset['Closest'].map(lambda x: int(x.lstrip('Distance From ')))
    dataset['color'] = dataset['Closest'].map(lambda x: colmap[x])
    
    return dataset

dataset = assignment(fish_data, centroids)


fig = plt.figure(figsize=(5,5))
plt.scatter(dataset['Height'],dataset['Weight'], color = dataset['color'], alpha=0.5, edgecolor='k')

for i in centroids.keys():
    plt.scatter(*centroids[i],color=colmap[i])
    
plt.xlim(int(min(fish_data['Height'])),int(max(fish_data['Height'])))
plt.ylim(int(min(fish_data['Weight'])),int(max(fish_data['Weight'])))
plt.show()
dataset.head()
```

```python
# Updating the cluster

import copy

old_centroids = copy.deepcopy(centroids)

def update(k):
    for i in centroids.keys():
        centroids[i][0] = np.mean(dataset[dataset['Closest'] == i]['Height'])
        centroids[i][1] = np.mean(dataset[dataset['Closest'] == i]['Weight'])
    return k

centroids = update(centroids)

fig = plt.figure(figsize=(5,5))
ax = plt.axes()
plt.scatter(dataset['Height'],dataset['Weight'], color = dataset['color'], alpha=0.5, edgecolor='k')

for i in centroids.keys():
    plt.scatter(*centroids[i],color=colmap[i])
    
plt.xlim(int(min(fish_data['Height'])),int(max(fish_data['Height'])))
plt.ylim(int(min(fish_data['Weight'])),int(max(fish_data['Weight'])))

for i in old_centroids.keys():
    old_x = old_centroids[i][0]
    old_y = old_centroids[i][1]
    dx = (centroids[i][0] - old_centroids[i][0])*0.75
    dy = (centroids[i][1] - old_centroids[i][1])*0.75
    ax.arrow(old_x, old_y, dx, dy, head_width=0.5, head_length=6, fc=colmap[i], ec=colmap[i])
    
    
plt.show()
```

```python
# Repeat Assignment Stage

dataset = assignment(dataset, centroids)

#plotting the results
fig = plt.figure(figsize=(5,5))
plt.scatter(dataset['Height'],dataset['Weight'], color = dataset['color'], alpha=0.5, edgecolor='k')

for i in centroids.keys():
    plt.scatter(*centroids[i],color=colmap[i])

plt.xlim(int(min(fish_data['Height'])),int(max(fish_data['Height'])))
plt.ylim(int(min(fish_data['Weight'])),int(max(fish_data['Weight'])))
plt.show()
```

```python
# Continue until all assigned categories don't change any more

while True:
    closest_centroids = dataset['Closest'].copy(deep=True)
    centroids = update(centroids)
    dataset = assignment(dataset, centroids)
    
    if closest_centroids.equals(dataset['Closest']):
        break
        
#plotting the results
fig = plt.figure(figsize=(5,5))
plt.scatter(dataset['Height'],dataset['Weight'], color = dataset['color'], alpha=0.5, edgecolor='k')

for i in centroids.keys():
    plt.scatter(*centroids[i],color=colmap[i])

plt.xlim(int(min(fish_data['Height'])),int(max(fish_data['Height'])))
plt.ylim(int(min(fish_data['Weight'])),int(max(fish_data['Weight'])))
plt.show()
```
