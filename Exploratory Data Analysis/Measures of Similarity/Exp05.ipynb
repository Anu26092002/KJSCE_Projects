<h1>1.	Identify the suitable attributes to apply the numeric similarity measures and write python code to calculate Euclidean, Manhattan similarity measures on it.</h1>

```python
# Function to calculate Euclidean Distance.
def euclideanDistance(row1, row2):
    sum = 0
    for i in range(len(row1)):
        sum += (row1[i] - row2[i])**2
    return sum**0.5

# Function to calculate Manhattan Distance.
def manhattanDistance(row1, row2):
    sum = 0
    for i in range(len(row1)):
        sum += abs(row1[i] - row2[i])
    return sum
```

```python
fish_data = fish_data.to_numpy()
fish_data
```

```python
print(euclideanDistance(fish_data[0],fish_data[1]))
```

```python
print(manhattanDistance(fish_data[0],fish_data[1]))
```

```python
def euclideanMatrix(dataset):
    n = np.shape(dataset)[0]
    matrix = [ [0]*n for i in range(n)]
    for i in range(n):
        for j in range(n):
            matrix[i][j] = euclideanDistance(dataset[i],dataset[j])

    return matrix
```

```python
matrix = euclideanMatrix(fish_data)
EuclideanMatrix = pd.DataFrame(matrix)
print("The Euclidean Distance Matrix is as followed : ")
EuclideanMatrix
```

```python
def manhattenMatrix(dataset):
    n = np.shape(dataset)[0]
    matrix = [ [0]*n for i in range(n)]
    for i in range(n):
        for j in range(n):
            matrix[i][j] = manhattanDistance(dataset[i],dataset[j])

    return matrix
```

```python
matrix = manhattenMatrix(fish_data)
ManhattenMatrix = pd.DataFrame(matrix)
print("The Manhatten Distance Matrix is as followed : ")
ManhattenMatrix
```

<h3>Reading the genomic sequence from the text file</h3>

```python
# Funtion to find the longest common subsequence by using Dynamic programming.
def longestCommonSubsequence(X, Y):
    
    # find the length of the strings
    m = len(X)
    n = len(Y)
  
    # declaring the array for storing the  values
    L = [[None]*(n + 1) for i in range(m + 1)]
  
    for i in range(m + 1):
        for j in range(n + 1):
            if i == 0 or j == 0 :
                L[i][j] = 0
            elif X[i-1] == Y[j-1]:
                L[i][j] = L[i-1][j-1]+1
            else:
                L[i][j] = max(L[i-1][j], L[i][j-1])

    return L[m][n]
```

```python
print('Longest Common Subsequence between str1 and str2 =',longestCommonSubsequence(str1, str2))
```

```python
def editDistance(X,Y):
    m = len(X)
    n = len(Y)
    
    array = [[0 for x in range(n + 1)] for x in range(m + 1)]
    
    for i in range(n+1):
        array[0][i] = i
    
    for i in range(m+1):
        array[i][0] = i
    
    for i in range(1,m+1):
        for j in range(1,n+1):
            if(X[i-1]==Y[j-1]):
                array[i][j] = array[i-1][j-1]
            else:
                array[i][j] = 1 + min(array[i][j-1],min(array[i-1][j],array[i-1][j-1]))
    return array[m][n]
```

```python
print('Edit Distance between str1 and str2 =',editDistance(str1, str2))
```
