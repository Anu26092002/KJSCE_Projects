<h1>1. Box Plot</h1>

```python
weight_data = np.array(fish_data['Weight'])
length1_data = np.array(fish_data['Length1'])
length2_data = np.array(fish_data['Length2'])
length3_data = np.array(fish_data['Length3'])
height_data = np.array(fish_data['Height'])
width_data = np.array(fish_data['Width'])
data = [ length1_data, length2_data, length3_data, height_data, width_data]

fig = mplt.figure(figsize =(10, 7))
ax = fig.add_subplot(111)
 
# Creating axes instance
bp = ax.boxplot(data, patch_artist = True, notch ='True', vert = 0)
 
colors = ['#00FF00','#FFCB00', '#CBCCAF', '#AAA000', '#00FFFF']
 
for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)

# Changing the color and style of the wiskers
for whisker in bp['whiskers']:
    whisker.set(color ='#8B008B',linewidth = 1.5,linestyle =":")
    
# Changing color and linewidth of caps
for cap in bp['caps']:
    cap.set(color ='#8B008B',linewidth = 2)
    
# Changing color and linewidth of medians
for median in bp['medians']:
    median.set(color ='red', linewidth = 2)
    
# Changing style of fliers
for flier in bp['fliers']:
    flier.set(marker ='D', color ='#e7298a', alpha = 0.5)
     
# x-axis labels
ax.set_yticklabels(['Length1','Length2', 'Length3', 'Height', 'Width'])
 
# Adding title
mplt.title("Box Plot of 'Length1','Length2', 'Length3', 'Height', 'Width' attributes of fishes in the dataset")
 
# Removing top axes and right axes
# ticks
ax.get_xaxis().tick_bottom()
ax.get_yaxis().tick_left()
     
# show plot
mplt.show()
```

<h1>2. Q-Q plot </h1>

```python
data = fish_data.sort_values(by=['Weight'], ascending=True).reset_index()
```

```python
data['Count'] = data.index + 1
```

```python
data.head()
```

```python
n_rows = data.shape[0]
n_rows
```

```python
data['Percentile Area'] = (data['Count'])/n_rows
data
```

```python
data['Z Theoritical'] = ndtri(data['Percentile Area'])
data
```

```python
data['Z Actual'] = (data['Weight']-data['Weight'].mean())/data['Weight'].std(ddof=0)
data
```

```python
mplt.scatter(data['Z Actual'],data['Z Theoritical'])
mplt.plot([-2,-1,0,1,2,3],[-2,-1,0,1,2,3], color="r")
mplt.xlabel('Data Quantiles') 
mplt.ylabel('Normal Quantiles') 

# displaying the title
mplt.title("Quantile-Quantile plot for distribution of fish weight")
  
mplt.show() 
```
