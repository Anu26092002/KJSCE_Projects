In the field of Machine Learning, the Perceptron is a Supervised Learning Algorithm for binary classifiers. The Perceptron Model implements the following function:

Perceptron networks come under single-layer feed-forward networks and are also called simple perceptrons. The perceptron network consists of three units, namely, sensory unit (input unit), associator unit (hidden unit), response unit (output unit). The sensory units are connected to associator units with fixed weights having values 1, 0 or -1, which are assigned at random.

```python
def customActivationFunction(x):
  return tf.cast(tf.greater(x, 0.8), dtype=tf.float32)
```

```python
''' 
    Defining the Class Model that will initialize the weights and the bias 
    and then calculate the sum by using the custom activation function. 
'''
class ModelwithCustomActivationFunction(object):

  def __init__(self):
    # Initialize the weights and the bias to random values
    self.W1 = tf.Variable(random.random())
    self.W2 = tf.Variable(random.random())
    self.b = tf.Variable(random.random())
  
  # Call Method returning the output of the Model.
  def __call__(self, x1,x2):
    self.K1 = self.W1 * x1 + self.W2 * x2 + self.b
    self.Output = customActivationFunction(self.K1)
    return(self.Output)

# Calling the object of the class
model0 = ModelwithCustomActivationFunction()
```

```python
def compute_error(target_y, predicted_y):
  error = tf.subtract(target_y, predicted_y)
  mse = tf.reduce_mean(tf.square(error))
  return mse
```

```python
def None_to_Zero(v):
   if v==None:
      v=0
   return v
```

```python
def train(model, X1, X2, Y, learning_rate, epochNo):
    with tf.GradientTape() as t:
      current_loss = compute_error(Y,model(X1,X2))
      print('Epoch Number: '+str(epochNo+1)+' Error: ')
      tf.print(current_loss)
      
      dW1,dW2,db=t.gradient(current_loss,[model.W1,model.W2,model.b])
      dW1=None_to_Zero(dW1) 
      dW2=None_to_Zero(dW2)
      db=None_to_Zero(db)
      
      model.W1.assign_sub(learning_rate * dW1)
      model.W2.assign_sub(learning_rate * dW2)
      model.b.assign_sub(learning_rate * db)
```

```python
def trainCustom(model, X1, X2, Y, learning_rate, epochNo):
    with tf.GradientTape() as t:
      current_loss = compute_error(Y,model(X1,X2))
      print('Epoch Number: '+str(epochNo+1)+' Error: ')
      tf.print(current_loss)
      
      dW1,dW2,db=t.gradient(current_loss,[model.W1,model.W2,model.b])
      dW1=None_to_Zero(dW1) 
      dW2=None_to_Zero(dW2)
      db=None_to_Zero(db)
      
      model.W1.assign_add(learning_rate * dW1)
      model.W2.assign_add(learning_rate * dW2)
      model.b.assign_add(learning_rate * db)
```

```python
def prepare_data(df, target_col="y"):
    X = df.drop(target_col, axis=1)
    
    y = df[target_col]
    
    return X, y 
```

```python
def display_plot(df, model, plot_dir="plots"):

  X, y = prepare_data(df_OR)

  def _create_base_plot(df):
    df.plot(kind="scatter", x="x1", y="x2", c="y", s=100, cmap="coolwarm")
    plt.axhline(y=0, color="black", linestyle="--", linewidth=1)
    plt.axvline(x=0, color="black", linestyle="--", linewidth=1)
          
    figure = plt.gcf()
    figure.set_size_inches(10, 8)
      
  def _plot_decision_regions(X, y, classifier, resolution=0.02):
    colors = ("cyan", "lightgreen")
    cmap = ListedColormap(colors)
          
    X = X.values 
    x1 = X[:, 0]
    x2 = X[:, 1]
          
    x1_min, x1_max = x1.min() - 1, x1.max() + 1 
    x2_min, x2_max = x2.min() - 1, x2.max() + 1
          
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution),np.arange(x2_min, x2_max, resolution))
    y_hat = model(xx1.ravel(), xx2.ravel()).numpy()
    y_hat = y_hat.reshape(xx1.shape)
          
    plt.contourf(xx1, xx2, y_hat, alpha=0.3, cmap=cmap)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())
          
    plt.plot()

  _create_base_plot(df)
  _plot_decision_regions(X, y, model)  
```
